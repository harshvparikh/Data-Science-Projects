## DECISION TREE CLASSIFIER
Create the Decision Tree classifier and visualize it graphically. If we feed any new data to this classifier, it should be able to predict the right class accordingly.
Dataset: "Iris.csv"

###### Importing the required libraries and the Iris dataset
Inpsect the data using the .head() method and the .shape attribute.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("Iris.csv")
print(data.head())
print(data.shape)
```

###### Check the classes to predict and features to be used

```python
print('Classes to predict: ',data['Species'].unique())
print('Features: ',data.columns[1:5].values)
```

###### Prepare the input and output values

```python
X = data.iloc[:,1:5].values
y = data.iloc[:,5].values
print(X.shape,y.shape)
```

###### Split the data, create the classifier and fit the model to our data
75% of the data becomes the training data and 25% of the data becomes the testing data.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 3)

classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

###### Visualize the decision tree

```python
from sklearn import tree
from sklearn import datasets
iris = datasets.load_iris()
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(classifier, 
                   feature_names=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'],  
                   class_names=['setosa', 'versicolor', 'virginica'],
                   filled=True)
```

###### Test the accuracy of our Decision Tree Classifier

```python
prediction = classifier.predict(X_test)
from sklearn import metrics
print("Accuracy of the Decision Tree: ",round(metrics.accuracy_score(prediction,y_test)*100,2),'%')
```
